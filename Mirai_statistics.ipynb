```python
df = pd.read_csv('data/Mirai_US_sales.csv')
df
```

```python
plt.figure(figsize=(10,6))

y1 = df['2016']
y2 = df['2017']
y3 = df['2018']
y4 = df['2019']
y5 = df['2020']
x = df['Month']
plt.xticks(rotation=-30)
#plt.xlim(1980, 2017)
#plt.scatter(x, y1)
plt.plot(x, y1, '-o')
#plt.scatter(x,y2)
plt.plot(x, y2, '-o')
#plt.scatter(x,y3)
plt.plot(x, y3, '-o')
#plt.scatter(x,y4)
plt.plot(x, y4, '-o')
#plt.scatter(x, y5)
plt.plot(x, y5, '-o')

plt.title('Toyota Mirai Monthly Sales', fontsize=16)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Month', fontsize=18)
plt.ylabel('Total Sold Number', fontsize=16)

plt.legend(loc="upper left", fontsize=15)


plt.show()
```

```python
X = ['2016', '2017', '2018', '2019', '2020']
Y = [1034, 1838, 1700, 1502, 499]

fig =  plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)


plt.plot(X, Y, c='b', linewidth = 5)
plt.scatter(X, Y, s=100, c = 'b')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

for i,j in zip(X,Y):
    ax.annotate(' '+str(j),xy=(i,j), fontsize=13)
    
plt.title('Toyota Mirai Yearly Sales', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Total Sold Number', fontsize=15)

plt.show()
```
